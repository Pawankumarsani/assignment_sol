### Q2. Join dataframes df1 and df2 by ‘fruit-pazham’ and ‘weight-kilo’.

```python
# Input
df1 = pd.DataFrame({'fruit': ['apple', 'banana', 'orange'] * 3,
                    'weight': ['high', 'medium', 'low'] * 3,
                    'price': np.random.randint(0, 15, 9)})

df2 = pd.DataFrame({'pazham': ['apple', 'orange', 'pine'] * 2,
                    'kilo': ['high', 'low'] * 3,
                    'price': np.random.randint(0, 15, 6)})
```

```python
# Solution
pd.merge(df1, df2, how='inner', left_on=['fruit', 'weight'], right_on=['pazham', 'kilo'], suffixes=['_left', '_right'])
```

### Q4. Create an empty dataframe and concatenate the Cars93 dataset for all Manufacturers one by one using a for loop

```python
# Input
df = pd.read_csv('https://raw.githubusercontent.com/selva86/datasets/master/Cars93_miss.csv')
```

```python
# Solution
colnames_df = df.columns

new_df = pd.DataFrame(columns = colnames_df)

for manu in df['Manufacturer'].unique():
    temp = df[df['Manufacturer'] == manu]
    temp.reset_index(drop = True, inplace = True)
    new_df = pd.concat([new_df, temp])
    new_df.reset_index(drop = True, inplace = True)
    
new_df
```
