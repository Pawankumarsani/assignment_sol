### Q1. Find out which vehicle type (`Type` column) has the maximum mean weight.

```python
# Input
df = pd.read_csv('https://raw.githubusercontent.com/selva86/datasets/master/Cars93_miss.csv')
```

```python
# Solution
df.groupby('Type').agg({'Weight': np.mean})
```

### Q2. Find out the standard deviation and mean price of vehicles for every vehicle `Type` - `DriveTrain` combination.

```python
# Input
df = pd.read_csv('https://raw.githubusercontent.com/selva86/datasets/master/Cars93_miss.csv')
df.head()
```

```python
# Solution
df.groupby(['Type', 'DriveTrain']).agg({'Price': [np.mean, np.std]})
```

```python
# Solution
df.groupby(['Type', 'DriveTrain']).agg({'Price': [np.mean, np.std]}).reset_index()
```

### Q4. Find the second largest value of 'taste' for 'banana' in dataframe df

```python
# Input
df = pd.DataFrame({'fruit': ['apple', 'banana', 'orange'] * 3,
                   'taste': np.random.rand(9),
                   'price': np.random.randint(0, 15, 9)})
```

```python
# Solution
df_grpd = df['taste'].groupby(df.fruit)
df_grpd.get_group('banana').sort_values().iloc[-2]
```

### Q5. Compute the mean price of every fruit, while keeping the fruit as another column instead of an index.

```python
# Input
df = pd.DataFrame({'fruit': ['apple', 'banana', 'orange'] * 3,
                   'rating': np.random.rand(9),
                   'price': np.random.randint(0, 15, 9)})
```

```python
# Solution
out = df.groupby('fruit', as_index=False)['price'].mean()
out
```

### Q6. Do a groupby operation on the Cars93 dataframe on `Type`. Iterate through each group and extract the names of top three cars which are Front wheel drive and has highest price.

```python
# Input
df = pd.read_csv('https://raw.githubusercontent.com/selva86/datasets/master/Cars93_miss.csv')
df.head()
```

```python
# Solution ---
# Group the dataframe by Pclass
for name, group in df.groupby('Type'): 
    print("Group name: ", name)
    group = group.sort_values('Price', ascending=False)
    print(group.loc[group.DriveTrain=="Front", ["Manufacturer", "Model", "Price"]].head(3), "\n\n")
```

### Q7. Add a new column in df having mean weight by `Type` of vehicle.

```python
# Input
df = pd.read_csv('https://raw.githubusercontent.com/selva86/datasets/master/Cars93_miss.csv')
```

```python
# Solution
df["Weight_Mean"] = df.groupby('Type')["Weight"].transform('mean')
df
```
