### Q1. Find out the distribution of vehicles between `Type` and `DriveTrain` of vehicles using cross tabulation. 

```python
# Input
df = pd.read_csv('https://raw.githubusercontent.com/selva86/datasets/master/Cars93_miss.csv')
```

```python
# Solution
rows = df['Type']
columns = df['DriveTrain']

pd.crosstab(rows, columns, rownames=['Type']) # count of vehicles

pd.crosstab(rows, columns, rownames=['Type'], normalize=True) # normalized values for comparision
```

### Q2. Compute a pivot showing the average price of people by:
1. Type vs DriveTrain
2. Type vs AirBags

```python
# Input
df = pd.read_csv('https://raw.githubusercontent.com/selva86/datasets/master/Cars93_miss.csv')
```

```python
# Solution 1
df_pivot1 = df.pivot_table(index='Type', columns='DriveTrain', values=['Price'], aggfunc=lambda x: np.mean(x))
df_pivot1
```

```python
# Solution 2
df_pivot2 = df.pivot_table(index='Type', columns='AirBags', values=['Price'], aggfunc=lambda x: np.mean(x))
df_pivot2
```

### Q3. Convert the car93 dataset from wide to long format:
- Store the Manufacturer and Model information as it. 
- Store the Type, DriveTrain and AirBags infromation in variable value pair

```python
# Input
df = pd.read_csv('https://raw.githubusercontent.com/selva86/datasets/master/Cars93_miss.csv')
df.head()
```

```python
# Solution
df_melted = pd.melt(df, id_vars=['Manufacturer', 'Model'], value_vars=['Type', 'DriveTrain', 'AirBags'])
df_melted
```

### Q4. Convert the long format data (output of previous question) to wide format 

```python
# Solution
df_wide = pd.pivot(df_melted, index=['Manufacturer', 'Model'], columns=['variable'], values='value')
df_wide
```
