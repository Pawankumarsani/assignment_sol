### Q1. Which manufacturer, model and type has the highest price in the Cars93 dataset.

```python
# Input
df = pd.read_csv('https://raw.githubusercontent.com/selva86/datasets/master/Cars93_miss.csv')
```

```python
# Solution
df.loc[df.Price == np.max(df.Price), ['Manufacturer', 'Model', 'Type']]
```

### Q2. Create a new column in Cars93 dataset which represents weight to highway mileage ration (Weight and MPG.highway columns).

```python
# Input
df = pd.read_csv('https://raw.githubusercontent.com/selva86/datasets/master/Cars93_miss.csv')
```

```python
# Solution 1
df['weight_mileage'] = df['Weight']/df['MPG.highway']
df.head()
```

### Q3. Print the 10th value in Price column in the Cars93 dataset and replace it with 50

```python
# Input
df = pd.read_csv('https://raw.githubusercontent.com/selva86/datasets/master/Cars93_miss.csv')
```

```python
# Solution 1
print(df['Price'][10]) # print 10th value

df['Price'][10] = 50 # replace 10th value

print(df['Price'][10]) # print 10th value
```

```python
# Solution 2

df = pd.read_csv('https://raw.githubusercontent.com/selva86/datasets/master/Cars93_miss.csv')
print(df.at[10, 'Price']) # print 10th value

df.at[10, 'Price'] = 50 # replace 10th value

print(df.at[10, 'Price']) # print 10th value
```

```python
# Solution 3

df = pd.read_csv('https://raw.githubusercontent.com/selva86/datasets/master/Cars93_miss.csv')
print(df.loc[10, 'Price']) # print 10th value

df.loc[10, 'Price'] = 50 # replace 10th value

print(df.loc[10, 'Price']) # print 10th value
```

### Q4. Filter the 'Manufacturer', 'Model' and 'Type' for every 20th row starting from 1st (row 0) in Cars93 dataset

```python
# Input
df = pd.read_csv('https://raw.githubusercontent.com/selva86/datasets/master/Cars93_miss.csv')
```

```python
# Solution
df.iloc[::20, :][['Manufacturer', 'Model', 'Type']]
```

### Q5. Filter out the car models which has price greater than 40 in the Cars93 dataset.

```python
# Input
df = pd.read_csv('https://raw.githubusercontent.com/selva86/datasets/master/Cars93_miss.csv')
```

```python
# Solution
df.loc[df['Price'] > 40, :]
```

### Q6. Filter out the car models which has price greater than 40  and less than 60 in the Cars93 dataset.

```python
# Input
df = pd.read_csv('https://raw.githubusercontent.com/selva86/datasets/master/Cars93_miss.csv')
```

```python
# Solution

filter1 = df['Price'] > 40
filter2 = (df['Price'] < 60)

df.loc[(filter1 & filter2), :]
```

### Q8. Sort `Manufacturer` in alphabetical order but `MPG.highway` in descending order in Cars93 dataset

```python
# Input
df = pd.read_csv('https://raw.githubusercontent.com/selva86/datasets/master/Cars93_miss.csv')
```

```python
# Solution
df.sort_values(by=['Manufacturer', 'MPG.highway'], ascending=[True, False])
```
